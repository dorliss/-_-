```python
import time
import keras_cv
import keras
import matplotlib.pyplot as plt
```

```python
model = keras_cv.models.StableDiffusion(
 img_width=512, img_height=512, jit_compile=False
)
```

```python
images = model.text_to_image("Fantastic cat with wolf live in the wild forest")
```

```python
def plot_images(images):
  plt.figure(figsize=(10, 10))
  for i in range(len(images)):
    ax = plt.subplot(1, len(images), i + 1)
    plt.imshow(images[i])
    plt.axis("off")
plot_images(images)
```

```python
prompt = "Majestic phoenix soaring through a cosmic storm, ethereal wings ablaze with iridescent hues, surrounded by swirling galaxies and celestial phenomena, digital art, highly detailed, cinematic composition, dramatic lighting, ethereal, mystical"
images = model.text_to_image(prompt, batch_size=3)
plot_images(images)
```
